# Calculator

A basic interactive calculator that performs arithmetic operations on two numbers.

```python
n1 = int(input("Enter number 1: "))
n2 = int(input("Enter number 2: "))
operator = input("Enter operator(+, -, *, /, %, **): ")

if operator == "+":
    print(n1 + n2)
elif operator == "-":
    print(n1 - n2)
elif operator == "*":
    print(n1 * n2)
elif operator == "/":
    print("Error: Cannot divide by zero" if n2 == 0 else n1 / n2)
elif operator == "%":
    print("Error: Cannot mod by zero" if n2 == 0 else n1 % n2)
elif operator == "**":
    print(n1 ** n2)
else:
    print("Invalid operator")
```

## How to Use

1. Run the cell below
2. Enter the **first number**
3. Enter the **second number**
4. Enter an **operator** from the list below
5. The result will be displayed

## Supported Operations

| Operator | Operation | Example |
|----------|-----------|---------|
| `+` | Addition | 5 + 3 = 8 |
| `-` | Subtraction | 5 - 3 = 2 |
| `*` | Multiplication | 5 × 3 = 15 |
| `/` | Division | 6 ÷ 2 = 3.0 |
| `%` | Modulo (remainder) | 7 % 3 = 1 |
| `**` | Exponentiation | 2 ** 3 = 8 |

## Features

Error handling for division by zero  
Error handling for invalid operators  
Simple user input format
