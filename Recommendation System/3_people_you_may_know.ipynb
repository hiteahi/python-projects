# People You May Know (Friend Recommendations) using Only Python

Given social network data, find friends-of-friends suggestions based on mutual connections

```python
import json
```

## 1. Load JSON data(Users, Pages)
- **Users**: id, name, friends, likes pages
- **Pages**: page id, page name

```python
def data_load(fname):
    with open(fname) as file:
        return json.load(file)
data = data_load("big_data.json") # not too big data
```

## 2. Function: people_you_may_know(user_id, data)

- build user_friends mapping (id -> set(friends))
- identify direct friends
- find mutual candidates
- sort by mutual count

```python
def people_you_may_know(user_id, data):

    user_friends = {}
    for user in data['users']:
        # take user id and friends list and make it a set for easier operations
         user_friends[user['id']] = set(user['friends']) 
    
    if user_id not in user_friends:
        return [] # if user id is not in the data, return empty list
    
    direct_friends = user_friends[user_id] # get the direct friends of the user

    suggestions = {}

    for friend in direct_friends: # For all friends of friend

        # If mutual id is not the same user and not already a direct friend of user_id, then add to suggestions
        for mutual in user_friends[friend]:
            # Check if mutual friend is not the user itself and not already a direct friend
            if mutual != user_id and mutual not in direct_friends:
                # Count mutual friends
                suggestions[mutual] = suggestions.get(mutual, 0) + 1

    # Sort suggestions by number of mutual friends in descending order
    sorted_suggestions = sorted(suggestions.items(), key=lambda x: x[1], reverse=True)
    
    # Create a dictionary for user names
    user_names = {user['id']: user['name'] for user in data['users']}

    # Get user name and suggested friends names
    user_name = user_names.get(user_id, f"User {user_id}")
    suggested_names = [user_names.get(s_id, f"User {s_id}") for s_id, _ in sorted_suggestions]

    return f"Name: {user_name}\nSuggested Friends: {suggested_names}"
```

## Step 3: Test with Example

Let's test the recommendation system with a sample user.

```python
# Example usage
user_id = 1

recommendations = people_you_may_know(user_id, data)
print(f"People You May Know for User\n{recommendations}")
```

## Summary
- **Data Loading**: Loads user and friendship data from big_data.json
- **Core Function**: people_you_may_know(user_id, data) finds potential friends by analyzing friends-of-friends
- **Algorithm**: Counts mutual connections and sorts suggestions by relevance
- **Output**: Returns list of recommended user name and suggested friends names
