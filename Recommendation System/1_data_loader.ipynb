# Function for Load Json Data

```python
import json
```

```python
def data_load(fname):
    with open(fname) as file:
        data = json.load(file)
        return data

data = data_load("data.json")
data
```

## Function for Display Users Detials 

```python
def display_users(data):
    
    print("Users and thier Connections\n")
    for users in data["users"]:
        print(f'ID {users["id"]}. {users["name"]} like this pages {users["liked_pages"]} and friendship with {users["friends"]}')
    
    print("\nPages Infomition\n")
    for page in data["pages"]:
        print(f'Page ID {page["id"]}. The Name of Page is {page["name"]}.')

display_users(data)
```
