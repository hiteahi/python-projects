# Data Cleaning using Only Python

## Problems in data

- Users Without Name
- Duplicate Friends
- Inactivate Users
- Duplicate Pages 

```python
import json
# load the data
data = json.load(open('data.json'))
data
```

## Function for Data Cleaning

```python
def clean_data(data):

    # remove user  without name  
    data['users'] = [user for user in data['users'] if user['name'].strip()] # empty list is falsy

    # remove duplicate friends
    for user in data['users']:
        user['friends'] = list(set(user['friends'])) # set for uniqueness, then back to list

    # remove inactive users
    data['users'] = [user for user in data['users'] if user['friends'] or user['liked_pages']] # keep users with friends or liked pages

    # Remove duplicate pages
    u_pages = {}
    for page in data['pages']:
        u_pages[page['id']] = page # dict will overwrite duplicates, keeping last occurrence
    data['pages'] = list(u_pages.values()) # back to list

    # Return the cleaned data
    return data

# clean and save the data
data = clean_data(data)
json.dump(data, open('clean_data.json', 'w'))


# display the cleaned data
print("Data cleaned successfully!")
data
```
