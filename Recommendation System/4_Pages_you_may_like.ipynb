# Pages You May Like - (Pages Recommendations) using Only Python

This notebook implements a recommendation engine that suggests pages to users based on what their friends like.

## Step 1: Load Data

First, we import necessary libraries and load the JSON data containing users and pages information.

```python
import json

def data_load(fname):
    with open(fname) as file:
        return json.load(file)
data = data_load("big_data.json")
```

## Step 2: Define Recommendation Function

The `pages_you_may_like()` function:
- Takes a user ID and data as input
- Analyzes the pages liked by the user's direct friends
- Recommends pages the user hasn't liked yet
- Ranks suggestions by how many friends like each page

```python
def pages_you_may_like(user_id, data):

    user_friends = {}
    user_liked_pages = {}
    for user in data['users']:
        user_friends[user['id']] = set(user['friends'])
        user_liked_pages[user['id']] = set(user['liked_pages'])
    
    # Create dictionaries for names
    user_names = {user['id']: user['name'] for user in data['users']}
    page_names = {page['id']: page['name'] for page in data['pages']}
    
    if user_id not in user_friends:
        return []  # if user id is not in the data, return empty list
    
    direct_friends = user_friends[user_id]  # get the direct friends of the user
    user_pages = user_liked_pages[user_id]  # get the pages the user already likes

    suggestions = {}

    for friend in direct_friends:  # For each friend
        for page in user_liked_pages[friend]:  # For each page the friend likes
            if page not in user_pages:  # If user doesn't already like it
                suggestions[page] = suggestions.get(page, 0) + 1

    # Sort suggestions by number of friends who like the page in descending order
    sorted_suggestions = sorted(suggestions.items(), key=lambda x: x[1], reverse=True)
    
    # Get user name and suggested page names
    user_name = user_names.get(user_id, f"User {user_id}")
    suggested_page_names = [page_names.get(p_id, f"Page {p_id}") for p_id, _ in sorted_suggestions]

    return f"Name: {user_name}\nSuggested Pages: {suggested_page_names}"
```

## Step 3: Test with Example

Let's test the recommendation system with a sample user.

```python
# Example usage
user_id = 5
recommendations = pages_you_may_like(user_id, data)
print(f"Pages You May Like for User ID {user_id}:\n{recommendations}")
```

## Summary

- **Data Loading**: Loads user and page data from big_data.json
- **Core Function**: `pages_you_may_like(user_id, data)` recommends pages based on what friends like
- **Algorithm**: Counts friend recommendations for each page and sorts by relevance
- **Output**: Returns user name and list of suggested page names
